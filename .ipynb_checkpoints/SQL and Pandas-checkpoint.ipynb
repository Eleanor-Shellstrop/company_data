```python
# Missing Client Contacts

conn = sqlite3.connect('company.db')
c = conn.cursor()

c.execute('''
SELECT
    company_orders.company_name,
    contact_first_name,
    contact_last_name,
    order_amount
FROM
    company_orders
LEFT JOIN company_client_list ON company_client_list.company_name = company_orders.company_name
WHERE company_client_list.contact_last_name IS NULL;
''')

missing_contacts = []

for row in c.fetchall():
    missing_contacts.append(row)
    print(row)
    
conn.commit()
conn.close()
```

```python
# Mutual instances: left join

conn = sqlite3.connect('company.db')
c = conn.cursor()

c.execute('''
SELECT
    client_number,
    company_orders.company_name,
    contact_first_name,
    contact_last_name,
    order_amount
FROM
    company_orders
LEFT JOIN company_client_list ON company_client_list.company_name = company_orders.company_name;
''')

all_clients = []

for row in c.fetchall():
    all_clients.append(row)
    print(row)
    
conn.commit()
conn.close()
```

```python
df = pd.DataFrame(data=all_clients, columns=
                  [
                      'company_id',
                      'company_name',
                      'contact_first_name',
                      'contact_last_name',
                      'order_amount'
                  ])

df.head()
```

```python
df['order_amount'] = df['order_amount'].str.replace('$', '', regex=True)

df.head()
```

```python
df = df.astype({'order_amount': 'float'})

df.dtypes
```

```python
# Strip whitespace

df.columns = [x.strip() for x in df.columns]
```

```python
contact_totals = df.groupby(['company_name', 'order_amount']).sum.reset_index()

contact_totals
```
